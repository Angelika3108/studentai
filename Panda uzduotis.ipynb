• Nuskaitykite studentų duomenų failo studentai.csv duomenis ir suformuokite iš jų
panda data frame.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
studentai = pd.read_csv('Studentai_panda_uzd_3.csv')
studentai.head() # Jonas Jonaitis turejo P10  19? pakeiciau i 10
```

• Bendras studentų skaičius.


```python
# suskaiciavo kiek yra eiluciu
bendrai_studentu=studentai.count()['Pavarde']   
print(F'Bendras studentu skaicius: {bendrai_studentu}')
```

• Apsiskaičiuokite kiekvieno studento vidurkį ir naujame stulpelyje „Egzaminas“
suformuokite vidurkio įvertinimą - teigiamas >=5 studentui leidžiama laikyti
egzaminą, o je neigiamas <5 laikyti egzamino studentui neįleidžiama.

```python
studentai.fillna(0, inplace=True)    # uzpildo NaN su 0

studentai['Vidurkis'] = round(studentai.iloc[:, 2:].mean(axis=1),1) #naujas stulpelis su vidurkiu
studentai.head() 
```

```python
studentai['Egzaminas'] = studentai['Vidurkis'].apply(lambda x: 'leidziama' if x >= 5 else 'neleidziama')
studentai.head()

# def ivertinimas(x):
#     return 'leidziama' if x >= 5 else 'neleidziama'

# studentai['Egzaminas'] = studentai['Vidurkis'].apply(ivertinimas)
```

• Kurie studentai turi teigiamą/neigiamą vidurkį. Kiek tokių studentų yra.


```python
teigiamas=studentai.loc[studentai['Egzaminas'] == 'leidziama'] 
teigiamu_skaicius=teigiamas.count()['Pavarde'] 
print(F'Egzamina leidziama laikiti {teigiamu_skaicius} studentams')
teigiamas.head()
```

```python
neigiamas=studentai.loc[studentai['Egzaminas'] == 'neleidziama']
negiamas_skaicius=neigiamas.count()['Pavarde'] 
print(F'Egzamina neleidziama laikiti {negiamas_skaicius} studentams')
neigiamas.head()
```

• Kurie studentai turi tiksliai 10 vidurkį. Kiek tokių studentų yra.

```python
desimtukai=studentai.loc[studentai['Vidurkis'] == 10]
desimtuku_skaicius=desimtukai.count()['Pavarde'] 
print(F'Desimtuku skaicius: {desimtuku_skaicius}')
desimtukai
```

• Kiek studentų vidurkis yra [5–6), [6–7), [7–8), [8–9), [9–10].

```python
five_six=studentai.loc[(studentai['Vidurkis'] >= 5) & (studentai['Vidurkis'] < 6)].count()['Pavarde'] 
six_seven=studentai.loc[(studentai['Vidurkis'] >= 6) & (studentai['Vidurkis'] < 7)].count()['Pavarde'] 
seven_eight=studentai.loc[(studentai['Vidurkis'] >= 7) & (studentai['Vidurkis'] < 8)].count()['Pavarde'] 
eight_nine=studentai.loc[(studentai['Vidurkis'] >= 8) & (studentai['Vidurkis'] < 9)].count()['Pavarde'] 
nine_ten=studentai.loc[(studentai['Vidurkis'] >= 9) & (studentai['Vidurkis'] < 10)].count()['Pavarde'] 
print(f'Studentu su vidurkiu 5-6: {five_six}')
print(f'Studentu su vidurkiu 6-7: {six_seven}') 
print(f'Studentu su vidurkiu 7-8: {seven_eight}')
print(f'Studentu su vidurkiu 8-9: {eight_nine}')
print(f'Studentu su vidurkiu 9-10: {nine_ten}')
```

• Pasirinkte atitinkamus grafikus rezultatams atvaizduoti. Panaudoti daugiau nei vieną
grafiko tipą.

```python
bins=[5,6,7,8,9,10] #nustato intervalus
plt.hist(studentai['Vidurkis'], bins=bins, edgecolor='black', alpha=0.7, color='pink')
plt.xlabel('Vidurkis')
plt.ylabel('Studentai')
plt.title('Studentu pazymiu pasiskirstymas')
plt.grid(True, linestyle='--', alpha=0.2)
```

```python
egzaminas_counts = studentai['Egzaminas'].value_counts()

egzaminas_counts.plot(kind='pie',
    autopct='%.2f%%',   
    startangle=90,       # pasuka grafika 90 laipsniu
    colors=['orange', 'lightgreen'],  
    labels=egzaminas_counts.index,  
    legend=False,
    figsize=(6, 6)       
)

plt.title('Egzamino leidimo pasiskirstymas')
plt.ylabel('') 
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# vidurkiu linija nuo 1mo studento sarase iki 51mo
plt.plot(studentai.index, studentai['Vidurkis'], label='Vidurkis', color='blue', marker='o')

# slenkstis egzaminui praeiti
plt.axhline(y=5, color='red', linestyle='--', label='Egzamino slenkstis (5)')

# Labels and title
plt.xlabel('Studento numeris')
plt.ylabel('Pazyma')
plt.title('Studentu vidurkiai ir egzamino leidimas')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#prideda nauja indeksa surusiojant pagal vidurki
studentai_sorted = studentai.sort_values(by='Vidurkis').reset_index(drop=False) 
studentai.to_csv('Studentai_naujas.csv', index=False) 
studentai_sorted.head()
```

```python
plt.figure(figsize=(10, 5))

# vidurkiu linija nuo maziausio iki didziausio
plt.plot(studentai_sorted.index, studentai_sorted['Vidurkis'], label='Vidurkis', color='blue', marker='o')

# slenkstis egzaminui praeiti
plt.axhline(y=5, color='red', linestyle='--', label='Egzamino slenkstis (5)')

plt.xlabel('Studento numeris (vidurkio didejimo tvarka)')
plt.ylabel('Pazyma')
plt.title('Studentu vidurkiai ir egzamino leidimas')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
